*The size column has 3 unique values:*

*k (Kilobytes): 36114*

*Varies with device : 74587*


Let's convert the M and K units into bytes and then remove the M and K from the values and convert them into numeric data type.

making a new column called 'Size in Mb' which will have the size in MB

```python
# convert the size column to numeric 
""" 
Binary (Traditional/IEC) Definition
Kilobyte    KB: 1,024 bytes (2^10)
Megabyte	MB: 1,048,576 bytes (2^20)
"""
def con_size(size):
    if isinstance(size,str):
        size=size.replace(',','')
        if 'k' in size :
            return float(size.replace('k',''))*1024
        elif 'M' in size :
            return float(size.replace('M',''))*1024*1024
        elif 'Varies with device' in size:
            return np.nan
    return size
```

```python
#applying function to the size column 
df['size']=df['size'].apply(con_size)
```

# Rating

#### ***Top 5 categories with ratings***

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top 5 categories by number of apps
top5_categories = df['category'].value_counts().head(5).index

# Calculate average rating for these categories
avg_rating_cat = df[df['category'].isin(top5_categories)].groupby('category')['rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
ax = sns.barplot(x=avg_rating_cat.index, y=avg_rating_cat.values, palette="GnBu")

plt.title("Average Ratings of Top 5 Most Popular Categories", fontsize=16)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Average Rating", fontsize=12)
plt.xticks(rotation=45, ha='right')

# Display rating values on top of bars
for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}",
                (p.get_x() + p.get_width()/2, p.get_height()),
                ha='center', va='bottom', fontsize=11)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#### ***Visualizing top 5 categories apps names with ratings***

```python
# -------- STEP 1: Select Top 5 Categories by Number of Apps (value_counts) --------
top5_categories = df['category'].value_counts().head(5).index

# -------- STEP 2: Filter only those categories --------
df_top = df[df['category'].isin(top5_categories)]

# -------- STEP 3: Get Top 5 apps (by installs) inside those categories --------
top_apps_per_category = (
    df_top.sort_values('installs', ascending=False)
          .groupby('category')
          .head(5)
)

# -------- STEP 4: Plot Horizontal Subplots --------
fig, axes = plt.subplots(5, 1, figsize=(18, 25))
sns.set(style="whitegrid")

for i, category in enumerate(top5_categories):
    data = top_apps_per_category[top_apps_per_category['category'] == category]
    
    sns.barplot(
        y='app_name',
        x='installs',
        data=data,
        ax=axes[i],
        palette="viridis"
    )
    
    axes[i].set_title(f"{i+1}. {category} ‚Äî Top 5 Apps by Installs", fontsize=16)
    axes[i].set_xlabel("Installs", fontsize=13)
    axes[i].set_ylabel("App Name", fontsize=13)
    
    # ---- Annotate ratings on bars ----
    for p, rating in zip(axes[i].patches, data['rating']):
        width = p.get_width()
        axes[i].text(
            width + width * 0.01,
            p.get_y() + p.get_height() / 2,
            f"Rating: {rating:.2f}",
            va='center', fontsize=11, fontweight='bold'
        )

plt.tight_layout(pad=2.5)
fig.suptitle("Top 5 Categories (Highest Number of Apps) ‚Üí Top 5 Apps per Category\nRatings Annotated Horizontally",
             fontsize=20, y=1.04)
plt.show()
```
